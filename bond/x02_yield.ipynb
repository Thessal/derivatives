## Bond Yield

* Zero rates
* Par Yield
* Bond Yield

Assume that you are a CFO, then you may be want to estimate the financing cost per each quarter.

You need USD 100 now, so you decided to sell bond at 100.

The problem is, how much coupon $c$ is required to sell this bond to the market?

| Time | Cashflow | 
|------|----------|
| 0Q   | 100      |
| 1Q   | -c       |
| 2Q   | -c       |
| 3Q   | -c       |
| 4Q   | -100 - c |

So you go to the bank. Banker rates your credit, and offers this : 

| Time | Zero rate, (cont. comp.)  |
|------|---------------------------|
| 0Q   | N/A                       |
| 1Q   | y1                        |
| 2Q   | y2                        |
| 3Q   | y3                        |
| 4Q   | y4                        |


Then, You can calculate PV of your bond, and estimate the cash payment that makes your bond attractive to the market.

$ 100 = c e^{-0.25 y_1} + c e^{-0.5 y_2} + c e^{-0.75 y_3} + c e^{-1.0 y_4} + 100 e^{-y_4} $

$ c  = 100 \frac{1 - e^{-y_4}}{  e^{-0.25 y_1} + e^{-0.5 y_2} + e^{-0.75 y_3} + e^{-1.0 y_4} } $

Then, par yield(the coupon rate) is $\frac{4c}{100}$

```python
# Example in Hull's book

import pandas as pd 
import numpy as np 

df = pd.DataFrame({
    "maturity" : [0, 0.5, 1, 1.5, 2],
    "zero rate": [np.nan, 0.05, 0.058, 0.0640, 0.0680], # continuous compounding
})

principal = 100
freq = 2
discount = df.eval("exp(- `maturity` * `zero rate`)")
coupon_discount = discount.sum() # "Value of an annuity that pays one dollar on each coupon payment date"
principal_discount = discount.iloc[-1] # "PV of one dollar received at the maturity of the bond"
coupon_rate = freq * (1 - principal_discount) / (coupon_discount)

print("Par yield = ", coupon_rate * 100, "%")
```
